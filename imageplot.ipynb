```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('left002012.png')

# Load the (x, y, z) coordinates from CSV
df = pd.read_csv('GroundTruth.csv')
x_coords = df['center_x'].values
y_coords = df['center_y'].values
# z_coords = df['z'].values

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the coordinates on the image
ax.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax.plot(x_coords, y_coords,  c='green')  # Adjust size and color as needed

# Set axis limits and aspect ratio
ax.set_xlim(0, image.shape[1])
# ax.set_ylim(0, image.shape[0])
ax.set_aspect('equal')

# Show the image with plotted points
plt.show()
```
